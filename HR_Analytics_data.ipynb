### Total Employee

```python
Total_emp = df['EmpID'].value_counts() 
Total_employee = len(Total_emp)
Total_employee
```

### Total Attrition

```python
Total_Attrition = df[df['Attrition']=='Yes']['Attrition'].count()
Total_Attrition
```

### Attrition Rate

```python
Attrition_rate = round((Total_Attrition/Total_employee)*100,1)
Attrition_rate
```

### Average Age 

```python
Average_Age = round(df['Age'].mean())
Average_Age
```

### Average Salary

```python
Avg_Age = df['MonthlyIncome'].mean()
Average_Age = Avg_Age.astype(int)
Average_Age
```

### Average Year at Company

```python
Avg_year_at_company = (df['YearsAtCompany'].mean()).astype(int)
Avg_year_at_company
```

# Exploratory Data Analysis

```python
Attrition = df[df['Attrition']=='Yes']

ax = sns.countplot(x= 'Gender',data= Attrition)

for i in ax.containers:
    ax.bar_label(i)
```

```python
Attrition = df[df['Attrition']=='Yes']

ax = sns.countplot(x= 'AgeGroup',data= Attrition)

for i in ax.containers:
    ax.bar_label(i)
```

```python
Attrition = df[df['Attrition']=='Yes']

ax = sns.countplot(x= 'SalarySlab',data= Attrition)

for i in ax.containers:
    ax.bar_label(i)
```

```python
Attrition = df[df['Attrition']=='Yes']

ax = sns.countplot(x= 'JobRole',data= Attrition)
plt.figure(figsize=(8,4))
for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

```python
Attrition.groupby(['JobRole'])['JobSatisfaction'].count().sort_values(ascending=False).reset_index()
```

```python
Attrition = df[df['Attrition']=='Yes']

ax = sns.countplot(x= 'Department',data= Attrition)
plt.figure(figsize=(8,4))
for i in ax.containers:
    ax.bar_label(i)
```

```python
Year_at_company = Attrition.groupby('YearsAtCompany')['YearsAtCompany'].count()
plt.figure(figsize=(19,8))
plt.title("Attrition by years at company ", fontsize = 30)
plt.plot(Year_at_company )
plt.xlabel('Year',fontsize=20)
plt.ylabel('Attrition Count',fontsize=20)
plt.show()
```
